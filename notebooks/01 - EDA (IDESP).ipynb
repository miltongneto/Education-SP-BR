# Exploratory Data Analysis - IDESP
*Índice de Desenvolvimento da Educação do Estado de São Paulo (IDEPS)*

#### Imports

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Load data

```python
PATH = '../data/raw'
```

```python
data = pd.read_csv(os.path.join(PATH, 'IDESP por Escola - 2018.csv'), sep=';')
print(data.shape)
data.head()
```

## Analise da distribuicao do IDESP

```python
data['ANOS_INICIAIS'].describe()
```

```python
data['ANOS_FINAIS'].describe()
```

```python
fig, axs = plt.subplots(1,3, figsize=(18, 8), sharey=True)
for i,col in enumerate(['ANOS_INICIAIS', 'ANOS_FINAIS', 'ENSINO_MEDIO']):
    sns.boxplot(y=data[col], ax=axs[i])
    
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,3, figsize=(18, 8), sharey=True)
for i,col in enumerate(['ANOS_INICIAIS', 'ANOS_FINAIS', 'ENSINO_MEDIO']):
    data[col].plot(kind='hist', ax=axs[i])
    
plt.tight_layout()
```
